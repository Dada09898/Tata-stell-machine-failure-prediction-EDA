# **Feature Engineering**


* Machine behavior ko better capture karne ke liye new features create karna.

````python
#  Data Wrangling
# Creating Business-Oriented Engineered Features

try:

    # Create a working copy to avoid modifying original data
    df_engineered = df.copy()


    # Create Temperature-Stress Index
    # This captures how temperature interacts with rotational speed
    # Using 'air temperature' as 'temperature' and 'rotational speed'
    df_engineered["temperature_stress_index"] = (
        df_engineered["air temperature"] * df_engineered["rotational speed"]
    )


    # Create Load Efficiency Ratio
    # Measures torque output relative to rotational speed
    # Using 'torque' and 'rotational speed'
    df_engineered["load_efficiency_ratio"] = (
        df_engineered["torque"] / (df_engineered["rotational speed"] + 1) # Add 1 to avoid division by zero
    )


    # Create Temperature Difference
    # Difference between process temperature and air temperature
    df_engineered["temperature_difference"] = (
        df_engineered["process temperature"] - df_engineered["air temperature"]
    )


    # The 'vibration' and 'pressure' columns are not explicitly present in the dataset
    # (e.g., 'vibration_torque_ratio' and 'pressure_temperature_interaction' from original template)
    # If such features are to be engineered, a clear mapping to existing columns (like TWF, HDF, PWF, OSF, RNF) would be needed,
    # or they should be omitted if no direct sensor data is available.
    # For now, these are omitted due to lack of direct column equivalents.


    # Display newly created columns
    print("New Engineered Features Created Successfully ")
    display(df_engineered.head())

except Exception as e:
    print("Error during Feature Engineering:", e)
````

# Find exact value jiske baad failure probability sharply increase hoti hai.

````python
# Numeric Cut-Off Detection for Torque

try:

    df = train_df.copy()

    # Sort dataset by torque
    df_sorted = df.sort_values("Torque [Nm]")

    # Create rolling failure rate (window based smoothing)
    window_size = 200

    df_sorted["rolling_failure_rate"] = (
        df_sorted["Machine failure"]
        .rolling(window=window_size)
        .mean()
    )

    # Find torque value where failure rate increases sharply
    threshold_index = df_sorted["rolling_failure_rate"].idxmax()
    torque_threshold = df_sorted.loc[threshold_index, "Torque [Nm]"]

    print("Estimated High-Risk Torque Threshold:", round(torque_threshold, 2))

except Exception as e:
    print("Error in Numeric Threshold Detection:", e)
````

````python
# Create Risk Flag Based on Detected Threshold

df["high_torque_risk"] = (
    df["Torque [Nm]"] > torque_threshold
).astype(int)

print("High Torque Risk Flag Created ")
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

Hypothetical Statement 1: Machines that experience failure have a significantly higher process temperature than machines that do not fail.

Null Hypothesis (H0): There is no significant difference in the mean Process_temperature_K between machines that fail and machines that do not fail.

H0: ( \mu_{\text{failure}} = \mu_{\text{no_failure}} )

Alternate Hypothesis (H1): The mean Process_temperature_K for machines that fail is significantly higher than for machines that do not fail.

H1: ( \mu_{\text{failure}} > \mu_{\text{no_failure}} )

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Hypothetical Statement 1:** Machines that experience failure have a significantly higher process temperature than machines that do not fail.

*   **Null Hypothesis (H0):** There is no significant difference in the mean `Process_temperature_K` between machines that fail and machines that do not fail.
    
    H0: \( \mu_{\text{failure}} = \mu_{\text{no_failure}} \)

*   **Alternate Hypothesis (H1):** The mean `Process_temperature_K` for machines that fail is significantly higher than for machines that do not fail.
    
    H1: \( \mu_{\text{failure}} > \mu_{\text{no_failure}} \)

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy import stats

# Separate the Process_temperature_K data into two groups based on 'Machine failure'
process_temp_no_failure = df[df['Machine failure'] == 0]['Process temperature [K]']
process_temp_failure = df[df['Machine failure'] == 1]['Process temperature [K]']

# Perform independent samples t-test (one-tailed)
# Since H1: mu_failure > mu_no_failure, we are interested in the right-tailed p-value.
# scipy.stats.ttest_ind returns a two-tailed p-value, so we divide by 2 for a one-tailed test
# if the t-statistic is positive, or (1 - p_value/2) if negative (though for > it should be positive)

t_statistic, p_value_two_tailed = stats.ttest_ind(process_temp_failure, process_temp_no_failure, equal_var=False)

# For a one-tailed test (H1: mu_failure > mu_no_failure):
# If t_statistic is positive, p_value_one_tailed = p_value_two_tailed / 2
# If t_statistic is negative, p_value_one_tailed = 1 - (p_value_two_tailed / 2)
# Since we expect mu_failure > mu_no_failure, we anticipate a positive t-statistic.

if t_statistic > 0:
    p_value_one_tailed = p_value_two_tailed / 2
else:
    p_value_one_tailed = 1 - (p_value_two_tailed / 2)

print(f"T-statistic: {t_statistic:.4f}")
print(f"One-tailed P-value (H1: mu_failure > mu_no_failure): {p_value_one_tailed:.4f}")
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.

An **Independent Samples t-test (Welch's t-test)** was performed to obtain the P-value. Specifically, a one-tailed version of the test was used, as our alternative hypothesis specifies a directional difference (mean process temperature for failed machines is *greater* than for non-failed machines).

##### Why did you choose the specific statistical test?

Answer Here.

The independent samples t-test was chosen because:
1.  **Nature of Variables**: We are comparing a continuous numerical variable (`Process_temperature_K`) across two independent categorical groups (`Machine failure` = 0 and `Machine failure` = 1).
2.  **Independence of Groups**: The two groups (failed vs. non-failed machines) are independent; the condition of one does not affect the other.
3.  **One-Tailed Hypothesis**: Our hypothesis (H1) is directional, stating that one mean is *greater* than the other, making a one-tailed test appropriate.
4.  **Unequal Variances**: Welch's t-test (by setting `equal_var=False`) is robust to unequal variances between the two groups, which is a common assumption violation in real-world data and a safer choice when group variances are unknown or likely different.

**Interpretation of P-Value and Conclusion:**

*   **T-statistic**: 12.5912
*   **One-tailed P-value**: 0.0000 (which indicates a value extremely close to zero, typically < 0.0001)

Given a significance level (alpha) of 0.05, our P-value (0.0000) is substantially less than alpha. This means there is extremely strong evidence against the null hypothesis.

**Conclusion:** We **reject the null hypothesis** (H0).

This implies that there is a statistically significant difference in the mean `Process_temperature_K` between machines that fail and machines that do not fail. Specifically, the data provides strong evidence to support the alternative hypothesis (H1) that the mean `Process_temperature_K` for machines that fail is significantly higher than for machines that do not fail.

**Outlier Treatment Technique:**
For all identified numerical columns ('Air_temperature_K', 'Process_temperature_K', 'Rotational_speed_rpm', 'Torque_Nm', 'Tool_wear_min'), the chosen outlier treatment technique was **Capping (Winsorization)**. Specifically, values below the lower bound (`Q1 - 1.5 * IQR`) were replaced with the lower bound, and values above the upper bound (`Q3 + 1.5 * IQR`) were replaced with the upper bound. This was done using the `.clip()` method in pandas.

**Justification for Capping:**
1.  **Preservation of Data Structure:** Capping retains all data points, preventing any loss of information that would occur if outliers were simply removed. This is particularly important for this dataset where 'Machine failure' is an imbalanced target, and removing rows could disproportionately affect the minority class.
2.  **Mitigation of Extreme Values' Impact:** Extreme outliers can heavily influence statistical analyses and machine learning models, leading to biased estimates, increased variance, and reduced model performance (especially for models sensitive to scale, like linear models or distance-based algorithms). Capping reduces the impact of these extreme values without entirely discarding them.
3.  **Domain Knowledge Consideration:** For operational parameters like temperature, rotational speed, torque, and tool wear, it's plausible that extreme values represent legitimate, albeit rare, operational conditions or sensor readings rather than pure errors. Capping allows these extreme conditions to still be represented within a more reasonable range, reflecting the physical limits or operational boundaries a machine might encounter, rather than treating them as impossible. For instance, a very high temperature might indicate an overheating event that is critical for predicting failure, and simply removing it might remove valuable information. Capping ensures it's still present but within a more model-friendly range.
4.  **Simplicity and Interpretability:** The IQR method and capping are straightforward to implement and interpret, providing a clear boundary for what is considered an acceptable range for each feature.

**Summary of Treatment:**
*   **Air_temperature_K:** Outliers capped to the calculated lower and upper IQR bounds (293.95 K and 305.55 K, respectively).

**Technique Used to Handle Imbalance:**
**SMOTE (Synthetic Minority Over-sampling Technique)** was used to handle the class imbalance. It was applied exclusively to the **training data (`X_train`, `y_train`)** after the initial train-test split.

**Rationale for Choosing SMOTE:**
1.  **Avoids Information Loss:** Unlike undersampling techniques (which remove instances from the majority class), SMOTE does not discard any valuable data from the dataset. This is important, especially for a problem where the minority class (failures) is critical and rare, and each instance might carry significant information.
2.  **Generates Synthetic Samples:** SMOTE works by creating synthetic samples for the minority class. It identifies the K-nearest neighbors for each minority class instance, randomly selects one of these neighbors, and creates a new, synthetic instance somewhere along the line segment joining the original instance and the selected neighbor. This helps in:
    *   **Increasing Minority Class Representation:** It balances the class distribution, ensuring that the model has enough examples of the minority class to learn from effectively.
    *   **Enriching Feature Space:** By generating synthetic samples rather than simply duplicating existing ones, SMOTE helps to expand the decision region of the minority class, making it less prone to overfitting than simple random oversampling.
3.  **Improved Model Generalization:** For highly imbalanced datasets, models trained on imbalanced data tend to be biased towards the majority class, leading to poor performance (especially low recall or precision) on the minority class. By balancing the training data, SMOTE helps the model to learn the characteristics of the minority class better, leading to improved generalization and more robust predictions for machine failures.
4.  **Application on Training Data Only:** SMOTE was applied only to the training data. This is crucial to prevent data leakage. If SMOTE were applied to the entire dataset before splitting, the test set would contain synthetic samples that are related to the training samples, leading to an artificially inflated performance evaluation.

After applying SMOTE, the training data (`y_train_resampled`) now has an equal number of instances for both classes (50% each), providing a balanced dataset for model training.

**Logistic Regression Model Explanation**

**What is Logistic Regression?**
Logistic Regression is a statistical model used for binary classification problems, meaning it predicts the probability of a binary outcome (e.g., 0 or 1, True or False, Failure or No Failure). Despite its name, it is a classification algorithm rather than a regression algorithm. It works by estimating the probability that a given instance belongs to a particular class.

**How it Works: The Sigmoid Function**
At its core, Logistic Regression uses the sigmoid (or logistic) function to map any real-valued number into a probability between 0 and 1. The sigmoid function is defined as:

`P(Y=1|X) = 1 / (1 + e^-(β0 + β1X1 + ... + βnXn))`

Where:
*   `P(Y=1|X)` is the probability that the dependent variable `Y` is 1, given the independent variables `X`.
*   `e` is the base of the natural logarithm.
*   `β0` is the intercept, and `β1...βn` are the coefficients for the independent variables `X1...Xn`.

The model takes a linear combination of input features (like in linear regression) and then applies the sigmoid function to this result. If the output probability is above a certain threshold (commonly 0.5), the instance is classified as the positive class (e.g., 'Failure'); otherwise, it's classified as the negative class ('No Failure').

**Strengths of Logistic Regression:**
1.  **Interpretability**: The coefficients of the logistic regression model can be interpreted as the change in the log-odds of the outcome for a one-unit change in the predictor variable. This makes it relatively easy to understand the impact of each feature on the probability of the outcome.
2.  **Efficiency and Simplicity**: It's computationally efficient and relatively easy to implement and train, especially for linearly separable data.
3.  **Output Probabilities**: It provides well-calibrated probabilities as output, which can be useful for ranking predictions or making decisions based on uncertainty.
4.  **Baseline Model**: Often serves as a good baseline model for classification tasks due to its simplicity and effectiveness.

**Limitations of Logistic Regression:**
1.  **Assumption of Linearity**: It assumes a linear relationship between the independent variables and the log-odds of the dependent variable. If this assumption is violated, the model's performance can suffer.
2.  **Sensitivity to Multicollinearity**: Like linear regression, it can be sensitive to multicollinearity (high correlation between independent variables), which can make coefficient estimates unstable and difficult to interpret.
3.  **Requires Feature Scaling**: For optimal performance and faster convergence during optimization, numerical features usually need to be scaled.
4.  **Cannot Handle Complex Relationships**: It may not perform well on datasets where relationships between features and the target variable are highly non-linear or complex, as it struggles to capture interactions between features directly.

    *   **What it measures:** The F1-Score is the harmonic mean of Precision and Recall. It provides a single metric that balances the concerns of both False Positives and False Negatives.
    *   **Why it's critical:** While recall is the highest priority, an extremely high recall achieved at the cost of very low precision (many false alarms) can also be detrimental. A **False Positive (Type I error)** occurs when the model predicts a machine will fail, but it does not. High numbers of false positives lead to:
        *   **Wasted Resources:** Unnecessary maintenance checks, inspections, and potentially premature parts replacement.
        *   **Production Interruptions:** Stopping a healthy machine for inspection still incurs downtime costs.
        *   **Loss of Trust:** Maintenance teams may start distrusting the system if too many alerts are false.
    *   **Business Impact:** The F1-Score helps ensure that while the model is good at catching failures (high recall), it's not generating an overwhelming number of false alarms (decent precision). A good F1-Score suggests a practical and actionable system that balances the need to prevent failures with the need to efficiently allocate maintenance resources.

3.  **ROC AUC Score: Good for Discriminative Power**
    *   **What it measures:** The ROC AUC (Receiver Operating Characteristic Area Under the Curve) measures the model's ability to distinguish between the positive and negative classes across all possible classification thresholds. It is robust to class imbalance.
    *   **Why it's critical:** A high ROC AUC score indicates that the model has excellent discriminative power, meaning it can effectively rank positive instances (failures) higher than negative instances (no failures). This is valuable because it shows the model's general capability regardless of a specific threshold.
    *   **Business Impact:** A high ROC AUC provides flexibility. It signifies that there is potential to adjust the classification threshold to fine-tune the trade-off between precision and recall based on evolving business priorities (e.g., if a new policy emphasizes safety even more, the threshold can be adjusted for higher recall). It ensures that the model is making good predictions that can be leveraged effectively.

**Accuracy and Precision (Less Primary Focus, but still considered):**
*   **Accuracy:** While important for overall correctness, it can be misleading in highly imbalanced datasets. A model predicting 'no failure' for all instances would achieve very high accuracy but miss all actual failures. Therefore, it's not the primary metric.
*   **Precision:** Minimizing false alarms is important for cost efficiency, but in this context, the cost of a missed failure (False Negative) is generally much higher than the cost of a false alarm (False Positive). Thus, precision takes a backseat to recall, but the F1-Score helps to keep it in check.
